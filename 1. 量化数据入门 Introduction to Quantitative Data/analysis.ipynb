## 基本计算 Simple Calculations

### 收益率计算 calculate of returns

```python
import yfinance as yf  # 导入 yfinance 模块
import matplotlib.pyplot as plt  # 导入 matplotlib 模块

# 计算简单收益率的函数
# Function to calculate simple returns
def calculate_simple_returns(data):
    # 计算简单收益率并删除 NA 值
    # Calculate simple returns and drop NA values
    simple_returns = data['Close'].pct_change().dropna()
    return simple_returns

# 计算累计收益率的函数
# Function to calculate cumulative returns
def calculate_cumulative_returns(simple_returns):
    # 计算累计收益率
    # Calculate cumulative returns
    cumulative_returns = (1 + simple_returns).cumprod() - 1
    return cumulative_returns

# 计算纳斯达克指数的简单和累计收益率
# Calculate simple and cumulative returns for the NASDAQ index
nasdaq_returns = calculate_simple_returns(nasdaq_data)
nasdaq_cumulative_returns = calculate_cumulative_returns(nasdaq_returns)

# 计算恒生指数的简单和累计收益率
# Calculate simple and cumulative returns for the Hang Seng Index
hsi_returns = calculate_simple_returns(hsi_data)
hsi_cumulative_returns = calculate_cumulative_returns(hsi_returns)

# 计算上证综合指数的简单和累计收益率
# Calculate simple and cumulative returns for the Shanghai Composite Index
shanghai_returns = calculate_simple_returns(shanghai_data)
shanghai_cumulative_returns = calculate_cumulative_returns(shanghai_returns)

# 创建一个新图形并设置大小
# Create a new figure with a specified size
plt.figure(figsize=(10, 6))

# 绘制纳斯达克指数的累计收益率
# Plot NASDAQ cumulative returns
plt.plot(nasdaq_cumulative_returns.index, nasdaq_cumulative_returns, label="NASDAQ", color='blue')

# 绘制恒生指数的累计收益率
# Plot Hang Seng cumulative returns
plt.plot(hsi_cumulative_returns.index, hsi_cumulative_returns, label="HSI", color='green')

# 绘制上证综合指数的累计收益率
# Plot Shanghai Composite Index cumulative returns
plt.plot(shanghai_cumulative_returns.index, shanghai_cumulative_returns, label="SSE", color='red')

# 添加图例
# Add a legend
plt.legend()

# 添加图表标题和轴标签
# Add chart title and axis labels
plt.title("Cumulative Returns of NASDAQ, Hang Seng, and Shanghai Composite Index (2022-01-01 to Present)", fontsize=14)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Cumulative Return", fontsize=12)

# 显示网格线
# Display grid lines
plt.grid(True)

# 显示图表
# Show the plot
plt.show()
```

### 标记买卖点 Mark Buy/Sell Points

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patheffects as path_effects

# 获取股票数据
# Get stock data
symbol = "600519.SS"
start_date = "2020-01-01"

data = yf.download(symbol, start=start_date)

# 计算移动平均
# Calculate moving averages
data['SMA_50'] = data['Close'].rolling(window=50).mean()
data['SMA_200'] = data['Close'].rolling(window=200).mean()

# 初始化交叉信号列
# Initialize the signal column for crossovers
data['Signal'] = 0

# 计算交叉信号
# Calculate crossover signals
# 通过比较50日均线和200日均线，生成买卖信号：
# Generate buy/sell signals by comparing the 50-day and 200-day moving averages:
# 当50日均线高于200日均线时，Signal 设为 1，表示买入信号。
# When the 50-day SMA is above the 200-day SMA, set Signal to 1 (buy signal).
# 当50日均线低于200日均线时，Signal 设为 -1，表示卖出信号。
# When the 50-day SMA is below the 200-day SMA, set Signal to -1 (sell signal).
data.loc[data['SMA_50'] > data['SMA_200'], 'Signal'] = 1
data.loc[data['SMA_50'] < data['SMA_200'], 'Signal'] = -1

# 找到买入和卖出信号的点
# Identify buy and sell signal points
# 买入点：Signal 从 -1 或 0 变为 1
# Buy signal: Signal changes from -1 or 0 to 1
# 卖出点：Signal 从 1 变为 -1
# Sell signal: Signal changes from 1 to -1
data['Position'] = data['Signal'].diff()  # 计算信号的变化
# Calculate the change in signals

buy_signals = data[data['Position'] == 2].index  # 买入信号
# Buy signals
sell_signals = data[data['Position'] == -2].index  # 卖出信号
# Sell signals

# 计算每日收益率
# Calculate daily returns
data['Daily_Return'] = data['Close'].pct_change()

# 计算策略信号的收益率（shift(1) 是为了避免未来数据的偏差）
# Calculate the returns based on strategy signals (shift(1) to avoid look-ahead bias)
# 如果我们在 当天 生成买卖信号并且 立即 根据这个信号进行交易，我们实际上是在 使用未来数据 来进行决策。
# If we generate buy/sell signals on a given day and immediately trade based on those signals,
# we would be using future data to make decisions.
# 这在实际交易中是不可能的，因为你不可能在当天的收盘价确定之前知道当天的所有数据。
# This is not possible in real trading, as you cannot know the day's closing price or all data before the day ends.
# 因此，为了避免这种“未来数据偏差”（即在未来数据已知的情况下进行交易），我们必须确保：当天生成的信号 应该 在下一天 才能生效。
# Therefore, to avoid this "look-ahead bias" (i.e., trading with the knowledge of future data),
# we must ensure that signals generated on one day should only take effect on the next trading day.
data['Strategy_Return'] = data['Signal'].shift(1) * data['Daily_Return']

# 计算累计收益
# Calculate cumulative returns
data['Cumulative_Return'] = (1 + data['Strategy_Return']).cumprod()

# 绘制价格图和均线
# Plot the price chart and moving averages
plt.figure(figsize=(14, 8))
plt.plot(data['Close'], label='Close Price', color='black', alpha=0.6)
plt.plot(data['SMA_50'], label='50-Day SMA', color='orange', linestyle='--', alpha=0.8)
plt.plot(data['SMA_200'], label='200-Day SMA', color='purple', linestyle='--', alpha=0.8)

# 标记买入点和卖出点
# Mark buy and sell signals
for signal in buy_signals:
    plt.text(
        signal,
        data.loc[signal, 'Close'],
        'B',
        color='red',
        fontsize=12,
        ha='center',
        va='center',
        fontweight='bold',
        path_effects=[
            path_effects.withStroke(linewidth=3, foreground='white'),
            path_effects.Normal()
        ]
    )
for signal in sell_signals:
    plt.text(
        signal,
        data.loc[signal, 'Close'],
        'S',
        color='green',
        fontsize=12,
        ha='center',
        va='center',
        fontweight='bold',
        path_effects=[
            path_effects.withStroke(linewidth=3, foreground='white'),
            path_effects.Normal()
        ]
    )

# 给买入点和卖出点画竖线
# Draw vertical lines for buy and sell signals
for signal in buy_signals:
    plt.axvline(x=signal, color='black', linestyle='--', alpha=0.5)

for signal in sell_signals:
    plt.axvline(x=signal, color='black', linestyle='--', alpha=0.5)

# 图表设置
# Chart settings
plt.title(f"{symbol} Buy and Sell Signals")
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```
